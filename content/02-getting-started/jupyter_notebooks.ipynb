```python
import numpy as np

import random
SEED=42
np.random.seed(SEED)
random.seed(SEED)

print(np.random.rand())
print(random.random())
```

```python
import torch
import torchvision

MNIST_data=torchvision.datasets.FashionMNIST(
      root='data',
      train='True',
      download='True'
)
#Get image data
images=MNIST_data.data.numpy()
#Get lables data
targets=MNIST_data.targets.numpy()
```

```python
targets[3]
```

```python
MNIST_data.classes
class_dict={i: classname for i, classname in enumerate (MNIST_data.classes)}

labels=[class_dict[target] for target in targets]
labels=np.array(labels)
```

```python
print(f"class_dict:{class_dict}")
print(f"labels:{labels}")
```

```python
print(f'Number of sample: {images.shape[0]}')
print(f'Shape of images: {images.shape}')
print(f'Shape of targets: {targets.shape}')
print(f'Shape of labels: {labels.shape}')
print(f'Shape of single images: {images[0].shape}')
print(f'Data type of an image: {images.dtype}')
print(f'Type of an images: {type(images[0])}')
```

```python
import pandas as pd
df=pd.DataFrame({
    "image": list(images),
    "label":labels
})
```

```python
print("Shape of the df: ", df.shape)
print("Column names of the df: ", df.columns.tolist())
```

```python
label_distribution=df['label'].value_counts()
label_distribution
```

```python
is_balanced=min(label_distribution)/max(label_distribution)>=0.9

print(f"Is the dataset balanced? {is_balanced}")
```
